```python
import json
```

```python
award_dict = json.load(open('award_2017.json','r'))
print(award_dict)
```

```python
team_dict = json.load(open('teams_2017_basic.json','r'))
```

```python
award_dict_keys = list(award_dict.keys())
print(award_dict_keys)
team_dict_keys = list(team_dict.keys())
```

```python
for team_name in award_dict_keys:
    for team_id in team_dict_keys:
        if team_dict[team_id]['team_name'] in team_name:
            team_dict[team_id]['award'] = award_dict[team_name]['award']
            team_dict[team_id]['medal'] = award_dict[team_name]['medal']
```

```python
print(team_dict[team_dict_keys[0]]['team_name'])
print(award_dict_keys[0])
```
