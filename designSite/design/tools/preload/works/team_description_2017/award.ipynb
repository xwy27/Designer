```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
res = requests.post('http://2017.igem.org/Competition/Results')
```

```python
with open('source.html','r') as f:
    soup = BeautifulSoup(f.read())
```

```python
#soup.find_all('b')[1].parent.parent.find_all('td')[-1].string
print('gold' in soup.find_all('b')[0].parent.parent.find('img')['src'])
#print(soup.find_all('b'))
```

```python
award_dict = dict()
```

```python
for index in range(len(soup.find_all('b'))):
    if index is 2 or index is 1:
        continue
    if soup.find_all('b')[index].string not in award_dict:
        award_dict[soup.find_all('b')[index].string] = {
                                                            'award':\
                                                            soup.find_all('b')[index]\
                                                            .parent.parent.find_all('td')[-1]\
                                                            .string\
                                                        }
    else:
        award_dict[soup.find_all('b')[index].string]['award'] = award_dict[soup.find_all('b')[index].string]['award']\
                                                            +';'+soup.find_all('b')[index]\
                                                            .parent.parent.find_all('td')\
                                                            [-1].string
    if 'gold' in soup.find_all('b')[index].parent.parent.find('img')['src']:
        award_dict[soup.find_all('b')[index].string]['medal'] = 'gold'
    elif 'silver' in soup.find_all('b')[index].parent.parent.find('img')['src']:
        award_dict[soup.find_all('b')[index].string]['medal'] = 'sivel'
    elif 'bronze' in soup.find_all('b')[index].parent.parent.find('img')['src']:
        award_dict[soup.find_all('b')[index].string]['medal'] = 'bronze'
    else:
        award_dict[soup.find_all('b')[index].string]['medal'] = ''
        
```

```python
json.dump(award_dict,open('award_2017.json','w'))
```

```python
for index in range(len(award_dict)):
    keys = list(award_dict.keys())
    if award_dict[keys[index]]['award'] is None:
        award_dict[keys[index]]['award'] = ''
```
